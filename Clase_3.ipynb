# 游늵 An치lisis Exploratorio de Datos (EDA) en Python 游냀

El An치lisis Exploratorio de Datos (EDA) es una t칠cnica fundamental en ciencia de datos que se utiliza para comprender mejor los datos antes de realizar cualquier modelado o an치lisis avanzado. En Python, hay varias bibliotecas populares como Pandas, Matplotlib y Seaborn que se utilizan com칰nmente para realizar EDA.

### Beneficios de las Visualizaciones en Python 游늵

1. **Comunicaci칩n Efectiva**: Las visualizaciones permiten comunicar informaci칩n de manera efectiva y comprensible, lo que facilita la interpretaci칩n de datos complejos.

2. **Toma de Decisiones Informada**: Al visualizar datos, se pueden identificar patrones, tendencias y relaciones que ayudan en la toma de decisiones informadas.

3. **Exploraci칩n de Datos**: Las visualizaciones ayudan a explorar los datos, descubrir insights ocultos y generar nuevas hip칩tesis.

4. **Presentaci칩n Atractiva**: Las visualizaciones atractivas hacen que los datos sean m치s accesibles y atractivos para una amplia audiencia.

### Librer칤as de Visualizaci칩n en Python y sus Usos 游늳

1. **Matplotlib**: 游늵 Ofrece una amplia variedad de gr치ficos est치ticos, desde simples hasta complejos, y es una excelente opci칩n para la generaci칩n de gr치ficos b치sicos.

2. **Seaborn**: 游 Se centra en la creaci칩n de gr치ficos estad칤sticos atractivos y est치 integrado con Pandas, lo que facilita la visualizaci칩n de datos almacenados en DataFrames.

3. **Plotly**: 游 Permite crear gr치ficos interactivos y din치micos, ideales para aplicaciones web y paneles interactivos.

## 游꿛 Paletas Secuenciales: para seaborn
- **游깯 viridis**: Una de las paletas predeterminadas, de tonos verdes a morados.
- **游눞 plasma**: De tonos morados a naranjas.
- **游댠 inferno**: De tonos oscuros a amarillos y naranjas brillantes.
- **游깲 magma**: De tonos oscuros a rosados y amarillos.
- **游릭 cividis**: Similar a viridis, pero optimizada para usuarios con deficiencias visuales de color.

## 游깽 Paletas Divergentes:
- **仇勇游 coolwarm**: De tonos fr칤os a c치lidos, excelente para visualizaci칩n de diferencias de temperatura o datos que tienen un punto medio.
- **游린游릳游릱 RdYlBu**: De rojo a amarillo y azul.
- **游 Spectral**: Paleta multicolor que va del rojo al azul pasando por el amarillo y verde.
- **游릮游릲 PuOr**: De morado a naranja.

## 游꿛 Paletas Categ칩ricas:
- **游댮游릭游릱 Set1, Set2, Set3**: Paletas categ칩ricas con colores vibrantes y distintos.
- **游 tab10, tab20**: Paletas categ칩ricas recomendadas para datos discretos, con una buena diferenciaci칩n entre colores.
- **游릮游리 Pastel1, Pastel2**: Colores suaves y apagados, 칰tiles para gr치ficos donde no se desea un contraste fuerte.
- **游깸 Dark2**: Colores oscuros, pero diferenciables entre s칤.

## 游깯 Paletas Perceptuales:
- **游 cubehelix**: Ideal para datos en escala continua y optimizada para una percepci칩n uniforme.
- **游 twilight**: De tonos suaves y oscuros en un ciclo.
- **游깾 twilight_shifted**: Similar a twilight, pero con un cambio en el rango de tonos.


````python
import pandas as pd # TAREA
import plotly.express as px

# Calculamos media y desviaci칩n est치ndar agrupadas por 'target'
stats_by_class = (
    data
    .groupby('target')['mean radius']
    .agg(['mean', 'std'])  # Calcula media y std (desviaci칩n est치ndar)
    .reset_index()
    .rename(columns={'mean': 'Mean Radius', 'std': 'Std Radius'})
)

# Creamos el gr치fico de barras con barras de error en Plotly Express
fig = px.bar(
    stats_by_class,
    x='target',          # Eje X
    y='Mean Radius',     # Eje Y (altura de la barra)
    error_y='Std Radius' # Agrega la desviaci칩n est치ndar como barra de error
)

fig.update_layout(
    title='Mean Radius by Class with Standard Deviation',
    xaxis_title='Class',
    yaxis_title='Mean Radius'
)

fig.show()
````

Gr치fico de Dispersi칩n de 'Petal_Length' vs 'Petal_Width'
游닇 Enunciado: Crea un gr치fico de dispersi칩n de 'petal_length' vs 'petal_width' con diferentes colores para cada especie en el conjunto de datos Iris.

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

especies = iris["species"].unique()
colores = ["red", "green", "blue"]

for especie, color in zip(especies, colores):
    datos = iris[iris["species"] == especie]
    plt.scatter(
        datos["petal length (cm)"],
        datos["petal width (cm)"],
        color=color,
        label=especie
    )

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width por Especie")
plt.legend()
plt.grid(True)
plt.show()
````

zip:recorre dos listas al mismo tiempo, una con las especies y otra con los colores. Es una forma compacta y ordenada de emparejar elementos.

Ejercicio 2: Histograma
Enunciado:
Crea un histograma para visualizar la distribuci칩n de la longitud del p칠talo (petal_length) para las tres especies de Iris. Usa colores para diferenciar las especies y ajusta el n칰mero de bins para mejorar la visualizaci칩n de los datos.

````python
import matplotlib.pyplot as plt

especies = iris["species"].unique()
colores = ["red", "green", "blue"]

plt.figure(figsize=(10, 6))

for especie, color in zip(especies, colores):
    datos = iris[iris["species"] == especie]
    plt.hist(
        datos["petal length (cm)"],
        bins=15,
        alpha=0.6,
        color=color,
        label=especie
    )

plt.xlabel("Petal Length (cm)")
plt.ylabel("Frecuencia")
plt.title("Distribuci칩n de la Longitud del P칠talo por Especie")
plt.legend(title="Species")
plt.grid(axis="y", alpha=0.3)
plt.show()
````

Crea un gr치fico de viol칤n (violin plot) para visualizar la distribuci칩n de la anchura del p칠talo (petal_width) en cada especie de Iris. Este gr치fico debe mostrar tanto la densidad como los valores individuales, lo que permite ver c칩mo se distribuyen los datos en cada grupo. - investigar teoria
