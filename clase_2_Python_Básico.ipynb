Problema 1: üé≤ Dados Divertidos üé≤
Enunciado del Problema: Queremos simular el lanzamiento de dos dados y contar cu√°ntas veces obtenemos una suma par.

````python
import random

def lanzar_dados_y_contar_suma_par(num_lanzamientos):
    """
    Simula el lanzamiento de dos dados y cuenta cu√°ntas veces obtenemos una suma par.

    Parameters:
    num_lanzamientos: El n√∫mero de lanzamientos de dados a simular.

    Returns:
    La cantidad de veces que la suma de los dados fue par.
    """
    suma_par = 0  # Inicializamos un contador para la suma par
    for _ in range(num_lanzamientos):  # Iteramos sobre el n√∫mero de lanzamientos
        dado1 = random.randint(1, 6)  # Generamos el resultado del primer dado
        dado2 = random.randint(1, 6)  # Generamos el resultado del segundo dado
        suma = dado1 + dado2  # Calculamos la suma de los resultados
        if suma % 2 == 0:  # Verificamos si la suma es par
            suma_par += 1  # Incrementamos el contador si la suma es par
    return suma_par  # Devolvemos la cantidad de sumas pares

# Simulamos 100 lanzamientos y contamos la cantidad de veces que la suma es par
num_lanzamientos = 100
resultado = lanzar_dados_y_contar_suma_par(num_lanzamientos)
print(f"De {num_lanzamientos} lanzamientos, {resultado} veces la suma fue par.")
````

Problema del M√°ximo N√∫mero üöÄ
Enunciado del Problema: Queremos encontrar el n√∫mero m√°s grande en una lista de n√∫meros. ¬°Tu tarea es escribir una funci√≥n que encuentre este n√∫mero m√°ximo! üìä

Instrucciones:

Escribe una funci√≥n llamada encontrar_maximo que tome una lista de n√∫meros como entrada.
Esta funci√≥n debe devolver el n√∫mero m√°s grande de la lista.
¬°Pon a prueba tu funci√≥n con una lista de n√∫meros y aseg√∫rate de que encuentre correctamente el m√°ximo! üßê
¬°Buena suerte! üçÄ

````python
def encontrar_maximo(lista): # Tarea
    """
    Encuentra el m√°ximo elemento en una lista de n√∫meros.

    Parameters:
    lista (list): La lista de n√∫meros.

    Returns:
    int or float: El valor m√°ximo en la lista.
    """
    if not lista:  # Verifica si la lista est√° vac√≠a
        return None  # Devuelve None si la lista est√° vac√≠a

    maximo = lista[0]  # Asigna el primer elemento de la lista como el m√°ximo inicialmente

    for num in lista:  # Itera sobre cada n√∫mero en la lista
        if num > maximo:  # Compara si el n√∫mero actual es mayor que el m√°ximo actual
            maximo = num  # Si es as√≠, actualiza el valor m√°ximo

    return maximo  # Devuelve el m√°ximo encontrado


# Ejemplo de uso
numeros = [12, 45, 67, 2300, 200, 34, 250]
maximo_numero = encontrar_maximo(numeros)
print(f"El m√°ximo n√∫mero en la lista es: {maximo_numero}")
````

##**zip en Python** / TAREA
La funci√≥n zip en Python toma iterables (objetos que se pueden iterar, como listas, tuplas, etc.) y los combina en una secuencia de tuplas. Cada tupla contiene elementos correspondientes de los iterables originales en el mismo √≠ndice. / **TAREA**
